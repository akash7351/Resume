```python


import os
import re
import PyPDF2
import pandas as pd

# Define a regular expression pattern to match skills (customize as needed)
skill_pattern = r"(\b[\w\s-]+ skills?\b)"

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_file):
    text = ""
    with open(pdf_file, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

# Directory containing the resume PDF files
resume_directory = r'C:\Users\lenovo\Downloads\archive\data\data\SALES'

# Create a DataFrame to store extracted details
resume_data = pd.DataFrame(columns=['Resume_ID', 'Category', 'Skills', 'Education'])

# Iterate through each resume file in the directory
for resume_file in os.listdir(resume_directory):
    if resume_file.endswith('.pdf'):
        resume_id = os.path.splitext(resume_file)[0]
        resume_text = extract_text_from_pdf(os.path.join(resume_directory, resume_file))
        
        # Extract Category, Skills, and Education (customize as needed)
        category = "TODO"  # Extract job category
        skills = re.findall(skill_pattern, resume_text, re.IGNORECASE)
        education = "TODO"  # Extract education details
        
        resume_data = resume_data.append({
            'Resume_ID': resume_id,
            'Category': category,
            'Skills': ', '.join(skills),
            'Education': education
        }, ignore_index=True)

# Now, resume_data DataFrame contains extracted details from the resumes.
```

```python
from datasets import load_dataset

# Load job descriptions dataset from Hugging Face
job_descriptions_dataset = load_dataset("jacob-hugging-face/job-descriptions")

# Example: Select 10-15 job descriptions
job_descriptions = job_descriptions_dataset['train']['job_description'][:15]

# Preprocess job descriptions (e.g., cleaning, tokenization)
```

```python
from transformers import DistilBertTokenizer, DistilBertModel
import torch

```

```python


# Initialize the DistilBERT tokenizer and model
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
model = DistilBertModel.from_pretrained("distilbert-base-uncased")

# Tokenize job descriptions (assuming job_descriptions is a list of strings)
job_desc_tokens = tokenizer(job_descriptions, padding=True, truncation=True, return_tensors="pt", max_length=512)

# Tokenize resume skills (assuming resume_data['Skills'] is a list of strings)
resume_tokens = tokenizer(resume_data['Skills'].tolist(), padding=True, truncation=True, return_tensors="pt", max_length=512)

# Preprocess data and convert to embeddings
with torch.no_grad():
    job_desc_embeddings = model(**job_desc_tokens).last_hidden_state.mean(dim=1)
    resume_embeddings = model(**resume_tokens).last_hidden_state.mean(dim=1)

# Convert embeddings to numpy arrays
job_desc_embeddings = job_desc_embeddings.cpu().numpy()
resume_embeddings = resume_embeddings.cpu().numpy()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity between job descriptions and resumes
similarity_scores = cosine_similarity(job_desc_embeddings, resume_embeddings)

# Rank candidates based on similarity scores for each job description
top_candidates_per_job = {}
for i, job_description in enumerate(job_descriptions):
    top_candidate_indices = similarity_scores[i].argsort()[::-1][:5]  # Get the top 5 candidates
    top_candidates_per_job[job_description] = [(resume_data.iloc[idx]['Resume_ID'], similarity_scores[i][idx]) for idx in top_candidate_indices]

# Now, top_candidates_per_job contains the top 5 candidates for each job description based on similarity scores.
```
